```python
patt = [", The", ", A", ", An"]
x= df['Title'].str.endswith(tuple(patt))
p_titles = df[x]
p_titles
```

```python
from dateutil.parser import parse

def parse_date(date_str):
    """Parse date string with multiple possible formats"""
    if pd.isna(date_str):
        return pd.NaT
    
    formats = [
        '%Y-%m-%d',   
        '%d-%m-%Y',   
        '%m-%d-%Y',    
    ]
    
    for fmt in formats:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except ValueError:
            continue
    
    
    try:
        return pd.to_datetime(parse(date_str))
    except:
        return pd.NaT


df['Publish_date'] = df['Publish_date'].apply(parse_date)
df
```

```python
df[df['Publish_date'].isna()]
```

```python
df['Publish_date'] = df.groupby('Title')['Publish_date'].transform(lambda x: x.fillna(x.dropna().iloc[0] if not x.dropna().empty else pd.NaT))
```

```python
df[df['Publish_date'].isna()]
```

```python
df = df.dropna(subset=['Publish_date'])
df[df['Publish_date'].isna()]
```

```python
df.drop(['Original_Publish_date'],axis=1,inplace=True)
df.columns
```

```python
df['Publish_date'] = df['Publish_date'].dt.date
df.head()
```

```python
df['Publish_date'] = df['Publish_date'].fillna('Unknown')

```

```python

df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    
    df['Is_Outlier'] = (df[col] < lower_bound) | (df[col] > upper_bound)
    outliers = df[df['Is_Outlier']]


    print(f"\nColumn: {col}")
    print(f"Lower bound: {lower_bound}")
    print(f"Upper bound: {upper_bound}")
    print(f"Number of outliers: {len(outliers)}")
    if not outliers.empty:
        print(outliers)


    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col], color='lightgrey', fliersize=0)
    sns.stripplot(
        x=df[col],
        hue=df['Is_Outlier'],
        palette={False: 'steelblue', True: 'red'},
        size=8,
        jitter=False
    )
    plt.axvline(x=upper_bound, color='red', linestyle='--', label=f'Upper Bound ({upper_bound:.2f})')
    plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound ({lower_bound:.2f})')
    plt.title(f"Outlier Visualization for {col}", fontsize=16)
    plt.xlabel(col)
    plt.legend(title="Outlier")
    plt.show()

df.drop(columns='Is_Outlier', inplace=True)

```
